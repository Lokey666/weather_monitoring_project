```python
import os
from dotenv import load_dotenv
import pandas as pd
from sqlalchemy import create_engine

load_dotenv()

DB_URL = os.getenv("DATABASE_URL")
engine = create_engine(DB_URL)

df_db = pd.read_sql("SELECT * FROM weather_data;", engine)

csv_path = r"C:\weather_monitoring_project\Data\live_raw.csv"

if not os.path.isfile(csv_path):
    df_db.to_csv(csv_path, mode='w', header=True, index=False)
    print("CSV created with headers.")
else:
    df_db.to_csv(csv_path, mode='a', header=False, index=False)
    print("New rows appended without header.")
```
