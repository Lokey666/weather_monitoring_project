```python
import os
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv

load_dotenv()

engine = create_engine(os.getenv("DATABASE_URL"))
csv_path = r"C:\weather_monitoring_project\Data\live_raw.csv"

# Fetch DB data
df_db = pd.read_sql("SELECT * FROM weather_data", engine)

# If CSV doesn't exist â†’ create it
if not os.path.isfile(csv_path):
    df_db.to_csv(csv_path, index=False)
    print("CSV created.")
else:
    # Load CSV
    df_csv = pd.read_csv(csv_path)

    # Make a unique key (timestamp + city)
    key_db = df_db['timestamp'].astype(str) + df_db['city']
    key_csv = df_csv['timestamp'].astype(str) + df_csv['city']

    # Filter only new rows
    df_new = df_db[~key_db.isin(key_csv)]

    # Append only new rows
    df_new.to_csv(csv_path, mode='a', header=False, index=False)
    print(f"Appended {len(df_new)} new rows.")
```
