```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os


raw_csv = r"C:\weather_monitoring_project\Data\live_raw.csv"

df = pd.read_csv(raw_csv)
df
```

```python
df.info()
df.describe()
```

```python
df['timestamp'] = (
    pd.to_datetime(df['timestamp'], utc=True, errors='coerce')
        .dt.tz_convert('Asia/Kolkata')
)
```

```python
df.isna().sum()
```

```python
weather_cols = ['temperature','humidity','wind_speed']
df[weather_cols] = df[weather_cols].interpolate(method='linear')
```

```python
poll_cols = ['pm10','pm2_5','nitrogen_dioxide','ozone']
df[poll_cols] = df[poll_cols].ffill()
df = df.bfill()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])

df['hour'] = df['timestamp'].dt.hour
df['day'] = df['timestamp'].dt.day
df['month'] = df['timestamp'].dt.month
df['weekday'] = df['timestamp'].dt.day_name()
```
